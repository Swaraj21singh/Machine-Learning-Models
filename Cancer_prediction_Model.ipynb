```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.neighbors import KNeighborsClassifier as Knn
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,r2_score
```

```python
# Cancer Prediction

data=pd.read_csv(r"D:\ducat\python\machine learning\global_cancer_patients_2015_2024.csv")
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
le=LabelEncoder()
#df["Gender_Encoded"]=le.fit_transform(df["Gender"])
#df["Cancer_Type_Encoded"]=le.fit_transform(df["Cancer_Type"])
df["Cancer_Stage_Encoded"]=le.fit_transform(df["Cancer_Stage"])
df
```

```python
X=df[["Age","Genetic_Risk","Air_Pollution","Alcohol_Use","Smoking","Obesity_Level","Treatment_Cost_USD","Survival_Years"]]
y=df["Cancer_Stage"]
z=df["Target_Severity_Score"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.001, random_state=42)
```

```python
log=LogisticRegression()
log.fit(X_train,y_train)
y_pred=log.predict(X_test)
y_pred
```

```python
X_test["prediction"]=y_pred
```

```python
X_test
```

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy of Logistic Regression:",accuracy_score(y_test,y_pred))
```

```python
X_test=X_test.drop("prediction",axis=1)
```

```python
## Knn Classifier
knn=Knn(n_neighbors=3)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
```

```python
y_pred
```

```python
X_test["prediction"]=y_pred
X_test
```

```python

print(f"Accuracy of Knn Classifier:",accuracy_score(y_test,y_pred))
```

```python
X_test=X_test.drop("prediction",axis=1)
```
